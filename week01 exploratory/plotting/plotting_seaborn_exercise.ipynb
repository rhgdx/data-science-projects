```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
plt.rcParams["figure.figsize"] = (12,7)
```

```python
penguins = pd.read_csv('./plotting_data/penguins_simple.csv', sep=';') #read penguins_simple.csv
```

[Seaborn Documentation](https://seaborn.pydata.org/examples/index.html)

--------------------------------

## Countplot 





* [Countplot](https://seaborn.pydata.org/generated/seaborn.countplot.html)

```python

penguins.columns = [columns.lower() for columns in penguins.columns.values.tolist()]
penguins.columns = [columns.replace(" ", "_") for columns in penguins.columns.values.tolist()]
```

```python
penguins.head()
```

```python
sns.countplot(data=penguins,y="species", hue = "sex")
```

## Univariate density estimation;  KDE plot



* [KDE plot](https://seaborn.pydata.org/generated/seaborn.kdeplot.html)

```python
sns.kdeplot(data = penguins, x="culmen_depth_(mm)", y="culmen_length_(mm)", hue = "species", fill=False, common_norm=True, multiple="stack") #this got weird
```

## Violinplot



* [Violinplot](https://seaborn.pydata.org/generated/seaborn.violinplot.html)

```python
sns.violinplot(data = penguins, x="species", y="culmen_length_(mm)", hue="sex", split=True, )
```

## Heatmap

* [Heatmap](https://seaborn.pydata.org/generated/seaborn.heatmap.html?highlight=heatmap#seaborn.heatmap)

```python
df = penguins.corr()
```

```python
sns.heatmap(data=df, annot=True, robust=True, cbar=False) #this one you need numeric data, its a bit fussy
```
